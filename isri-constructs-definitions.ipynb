```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def put_to_df(url):
    # URL'den HTML verisini çek
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Sayfada tanımlı tüm Construct'lar için işlem yap
    constructs = soup.select('div.tabs__content.active')  # 'active' class'ına sahip 'div'leri seçiyoruz

    # DataFrame'e eklemek için boş liste
    rows = []

    # Her bir construct'ı işleme al
    for construct in constructs:
        # Construct Name (başlık)
        h3_tag = soup.find('h3', text=lambda t: t and "Construct:" in t)
        if h3_tag:
            construct_name = h3_tag.get_text(strip=True).replace("Construct:", "").strip()
        else:
            construct_name = 'N/A'

        # Definitions (Definition) almak
        paragraphs = construct.find_all('p')  # Birden fazla p etiketi olabilir
        for paragraph in paragraphs:
            # Definition kısmını bul
            definition = paragraph.find('i', string="Definition:")
            if definition:
                # Definition başlığından sonra gelen metni al
                definition_text = paragraph.get_text(strip=True).replace("Definition:", "").strip()

                # Theory/model kısmını bul
                theory_model = paragraph.find('i', string="Theory/model:")
                if theory_model and theory_model.next_sibling:
                    # Theory/model kısmından sonraki metni al
                    theory_model_text = theory_model.next_sibling.strip()
                else:
                    theory_model_text = 'N/A'

                # Reference kısmını bul
                reference = paragraph.find('i', string="Reference:")
                if reference:
                    # Reference kısmından sonraki metni al
                    reference_text = reference.find_next('a').get_text(strip=True) if reference.find_next('a') else 'N/A'
                else:
                    reference_text = 'N/A'

                # Verileri ekle
                rows.append({
                    'Construct Name': construct_name,
                    'Definition': definition_text,
                    'Theory/model': theory_model_text,
                    'Reference': reference_text
                })

    # Verileri birleştir
    new_df = pd.DataFrame(rows)
    return new_df
```

```python
# Key-value 쌍을 생성
con_dict = {}
for con in con_list:  # 
    key = con.replace(" ", "%20")  # 공백을 '%20'으로 변경
    con_dict[con] = key

# 생성된 key-value 쌍을 출력
for con, key in con_dict.items():
    print(f"{con}: {key}")  # 키와 값을 출력합니다.
```

```python
final_df = pd.DataFrame()
```

```python
for con, key in con_dict.items():
    temp_df = put_to_df("https://isri.sciencesphere.org/index.php?o=construct&c="+key)
    final_df = pd.concat([final_df, temp_df], ignore_index=True) 
```
